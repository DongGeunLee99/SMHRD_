### <h3>Lib , read</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as knn

# 모델 성능 평가 도구
from sklearn.metrics import accuracy_score
```

```python
# 데이터 불러오기 (준비)
data = pd.read_csv('./data/bmi_500.csv')
data
```

```python
# # 데이터 탐색
# data.head()
# data.tail()
# data.shape
# data.ndim # n차원
# data.info()
```

### <h3>데이터 전처리</h3>

```python
# 데이터 전처리 (분리)
data.head(2)
```

```python
X = data.loc[:,['Height', 'Weight']]    # 문제 데이터 : 키와 체중
y = data.loc[:,'Label']                 # 답 데이터 : 라벨

# 데이터 분리 이후 크기 확인 필수!!!!
print("X = {} \ny = {}".format(X.shape, y.shape))   
```

```python
# 학습지와 시험지로 행 분리
# 머신러닝 훈련 데이터셋
# 훈련 : 평가 = 7 : 3
# 500개의 행을 350:150으로 나누기
# X_train, X_test, y_train, y_test
```

```python
X_train, y_train, X_test, y_test = X[:350], y[:350], X[350:], y[350:]
print("X_train = {} \ty_train = {} \nX_test = {} \ty_test = {}"
        # .format(X_train.shape, y_train.shape, X_test.shape, y_test.shape))
        .format(len(X_train), len(y_train), len(X_test), len(y_test)))
```

### <h3>모델링</h3>
- 모델 객체 생성
- 모델 학습 진행
- 모델 평가 및 예측

```python
# knn 분류 모델 
knn_model = knn(n_neighbors=3) # 기본 값 5
```

```python
# 모델 학습
knn_model.fit(X_train, y_train)
```

```python
# 모델 예측
pre = knn_model.predict(X_test)
```

```python
# 모델 평가
accuracy_score(y_test, pre) # 실제 답과 데이터가 얼마나 일치한지
```
