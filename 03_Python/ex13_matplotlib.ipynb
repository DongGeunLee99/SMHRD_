```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 그래프에 한글을 표현하는 방법
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

```python

acc_data = pd.read_csv('./data/Traffic_Accident_2017.csv', encoding='euc-kr')
# acc_data.info()

car_acc = acc_data[acc_data['사고유형_대분류'] == '차대차']
day_car_acc = car_acc[car_acc['주야'] == '주간']
car_data = day_car_acc['요일'].value_counts()
result = car_data[['월', '화', '수', '목', '금', '토', '일']]

plt.figure(figsize=(3, 3))
plt.bar(result.index, result)
plt.title('요일별 주간 차대차 교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고건수')

# 그려지는 그래프의 범위를 수정 -> (시작, 끝)
plt.ylim(0, 200)
plt.show()
```

```python
score_list =   [['김미희', 89, 78, 99],
                ['서대희', 67, 88, 77],
                ['강예진', 100, 55, 96]]

for i in range(len(score_list)):
    score_list[i].append((score_list[i][1] + score_list[i][2] + score_list[i][3])//3)

for i in range(len(score_list)):
    if score_list[i][4] >= 90:
        score_list[i].append("A")
    elif score_list[i][4] >= 80:
        score_list[i].append("B")
    elif score_list[i][4] >=70:
        score_list[i].append("C")
    elif score_list[i][4] >=60:
        score_list[i].append("D")
    else:
        score_list[i].append("F")

print("="*45)
print("이름\tPython\tJava\tDB\t평균\t학점")
print("="*45)
for i in score_list:
    for j in range(len(i)):
        print(i[j],end='\t')
    print()
print("="*45)
py, ja, db = 0, 0, 0
for i in score_list:
    py += i[1]
    ja += i[2]
    db += i[3]
py //= len(score_list)
ja //= len(score_list)
db //= len(score_list)
print("평균\t{}\t{}\t{}".format(py,ja,db))
```

```python
import pandas as pd

score_list =   [['김미희', 89, 78, 99],
                ['서대희', 67, 88, 77],
                ['강예진', 100, 55, 96]]

idx = []
score = []
python=[]
java=[]
db=[]

for i in range(len(score_list)): 
    idx.append(score_list[i][0])
    python.append(score_list[i][1])
    java.append(score_list[i][2])
    db.append(score_list[i][3])

data = {'Python' : python, 'Java':java, 'DB':db}

df = pd.DataFrame(data, index=idx, )
df.index.name = '이름'

df['평균'] = df.mean(axis=1).apply(int)
ages = df['평균']
labels = ['F', 'D', 'C', 'B', 'A']
bins = [-1, 59, 69, 79, 89, 100]
cate = pd.cut(x=ages, bins=bins, labels=labels)
df['학점'] = cate

print("="*45)
print("이름\tPython\tJava\tDB\t평균\t학점")
print("="*45)

for i in range(len(df.values)):
    print(df.index[i], end='\t')
    for j in range(len(df.values[i])):
        print(df.values[i][j],end='\t')
    print()
print("="*45)
print("평균",end='\t')
avg = (df.loc[:,'Python':'DB'].mean(axis=0).apply(int))
for i in range(len(avg)): print(avg.values[i], end='\t')
```
