### 한솥도시락 페이지에 있는 전체 메뉴 및 가격 데이터 크롤링
- 라이브러리 로딩
- 페이지 전체 데이터를 수집하는 코드를 작성

```python
# 라이브러리

from selenium import webdriver as wb
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
```

```python
# 페이지 불러오기

s = Service(ChromeDriverManager().install())
driver = wb.Chrome(service=s)
driver.get('https://www.hsd.co.kr/menu/menu_list')
```

```python
# 메뉴명 수집 -> 객체화

suop = bs(driver.page_source, 'lxml')
suop
```

```python
menu = suop.select('h4.h.fz_03')
price = suop.select('div.item-price>strong')
```

```python
menulist = []
pricelist = []
for i in range(len(menu)):
    menulist.append(menu[i].text)
    pricelist.append(price[i].text)
print(menulist)
print(pricelist)
```

#### 더보기 버튼 누르기
- 한솥도시락 페이지는 더보기 버튼을 누르지 않으면 전체 메뉴가 보이지 않는다
- 더보기 버튼을 눌러서 모든 메뉴 데이터를 크롤링해보자

```python
# 더보기 버튼 위치 찾기
btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
btn_more.click()
```

### 예외처리 하는 방법
- 실행할 부분(try)은 모두 실행하고, 만약 오류가 나면 예외처리(except)

```python
# 코드 실행 시 에러가 발생하면 밑에 코드는 실행되지 않고 중간에 멈춘다.
# 멈추지 않고 계속 아래 코드를 실행할 수 있도록 예외처리를 진행
```

### time 라이브러리
- 코드 중간에 쉬는 시간을 주는 라이브러리

```python
import time
# time.sleep()
# 사람이 처리하는 것처럼 작업을 해야 서버에 부담이 가지 않는다.
```

```python
try:
    for i in range(10):
        btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
        btn_more.click()
        time.sleep(2)
except:
    print("완료")
```

```python
# 한솥 도시락 페이지의 모든 메뉴 명과 가격 데이터를 수집해보자
# 이 한 셀만 실행해도 모든 데이터가 수집되는 코드를 작성해보자

import time
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager


s = Service(ChromeDriverManager().install())
driver = wb.Chrome(service=s)
driver.get('https://www.hsd.co.kr/menu/menu_list')


try:
    for i in range(10):
        btn_more = driver.find_element(By.CLASS_NAME, 'c_05')
        btn_more.click()
        time.sleep(2)
except:
    print("완료")


suop = bs(driver.page_source, 'lxml')
menu = suop.select('h4.h.fz_03')
price = suop.select('div.item-price>strong')

menulist = []
pricelist = []
for i in range(len(menu)):
    menulist.append(menu[i].text)
    pricelist.append(price[i].text)

print(menulist)
print(pricelist)
print(len(menulist))
print(len(pricelist))

dic ={'메뉴' : menulist, '가격' : pricelist}
hansot = pd.DataFrame(dic)
hansot.set_index('메뉴', inplace=True)
hansot.to_csv('한솥.csv', encoding='utf-8')
hansot
```
